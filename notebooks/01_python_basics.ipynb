```python
def mean(lst):
    return sum(lst) / len(lst)
scores = [8,9,10]
print(mean(scores))
```

```python
def count_words(s):
    words = s.split()
    return len(words)

text = "toi dang hoc khoa hoc du lieu"
print(count_words(text))
```

```python
def char_freq(s):
    result = {}
    for ch in s:
        if ch == " ":
            continue
        result[ch] = result.get(ch, 0) + 1
    return result

print(char_freq("data  science"))
```
