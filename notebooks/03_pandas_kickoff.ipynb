```python
df = pd.DataFrame({
    "order_id": [1, 2, 3, 4],
    "customer": ["an", "binh", "an", "chi"],
    "amount": [200, 120, 450, 150],
    "region": ["north", "south", "north", "north"]
})

print(df)
print("shape:", df.shape)
```

```python
print(df["amount"])        # Series
print(type(df["amount"]))

print(df[["amount"]])      # DataFrame 1 cột
print(type(df[["amount"]]))
```

```python
high_amount = df[df["amount"] >= 200]
print(high_amount)
```

```python
print(df["amount"])        # Series
print(type(df["amount"]))

print(df[["amount"]])      # DataFrame 1 cột
print(type(df[["amount"]]))
```

```python
high_amount = df[df["amount"] >= 200]
print(high_amount)
```

```python
north_high = df[(df["region"] == "north") & (df["amount"] >= 200)]
print(north_high)
```
