```python
import pandas as pd
import numpy as np

def clean_orders(filepath='orders_dirty.csv'):
    df = pd.read_csv(filepath, sep=';', parse_dates=['Order Date'])
    
    # Làm sạch text
    text_cols = ['Customer', 'Region', 'Product']
    for col in text_cols:
        df[col] = df[col].astype(str).str.strip().str.lower()
    
    # Làm sạch số (có dấu phẩy, N/A, "")
    num_cols = ['Quantity', 'Revenue', 'Unit Price']
    for col in num_cols:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', ''), errors='coerce')
    
    # Xử lý NaN
    df = df.dropna(subset=['Customer', 'Revenue'])           # xóa dòng thiếu quan trọng
    df['Region'] = df['Region'].fillna('unknown')
    
    return df

# Dùng hàm
df_analysis = clean_orders('orders_dirty.csv')   # thay tên file của bạn
print("✅ Đã clean xong! Shape:", df_analysis.shape)
df_analysis.head(17)
```

```python
# Đếm số đơn theo Region
print(df_analysis['Region'].value_counts())

# Top 5 Region
print(df_analysis['Region'].value_counts().head(5))

# Tỉ lệ % 
print(df_analysis['Region'].value_counts(normalize=True).round(3) * 100)
```

```python
# Tổng hợp doanh thu + số lượng theo Region
summary = df_analysis.groupby('Region').agg({
    'Revenue': ['sum', 'mean', 'max'],
    'Quantity': 'sum',
    'Order Date': 'count'      # số đơn hàng
}).round(0)


sumary = df.analysis.groupby('Region')


print(summary)
```

```python
# Top 10 khách hàng có doanh thu cao nhất
top10 = df_analysis.groupby('Customer')['Revenue'].sum() \
                   .sort_values(ascending=False) \
                   .head(10)

print(top10)
```

```python
top_customers = df_analysis.groupby(['Region', 'Customer']).agg({
    'Revenue': 'sum',
    'Quantity': 'sum'
}).reset_index()

print(top_customers)
```

```python
# Top 10 khách hàng có doanh thu cao nhất
top10 = df_analysis.groupby('Customer')['Revenue'].sum().sort_values(ascending=False).reset_index().head(10)

print(top10)
```
