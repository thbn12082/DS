```python
import numpy as np

# Dữ liệu mẫu 4x3
data = np.array([[85, 92, 78],
                 [60, 88, 45],
                 [95, 70, 82],
                 [55, 91, 67]])

print("Dữ liệu gốc:")
print(data)
mask = data>= 80
print(mask)
print(len(mask))
print(data * mask)

data_nan = np.where( mask, data, np.nan)
print(data_nan)
```

```python
# Giả sử có dữ liệu nhiệt độ, muốn giữ chỉ nhiệt độ hợp lý (20-35°C)
nhiet_do = np.random.uniform(10, 40, (5, 7))   # 5 ngày x 7 giờ

mask_hop_ly = (nhiet_do >= 20) & (nhiet_do <= 35)

# Thay giá trị bất thường thành NaN
nhiet_do_sach = np.where(mask_hop_ly, nhiet_do, np.nan)

print("Nhiệt độ sau khi lọc:", nhiet_do_sach)
```

```python
a = np.zeros((2, 3))
b = np.ones((2, 2))
c = np.arange(0, 10, 2)   # 0,2,4,6,8
d = np.arange(1, 13).reshape(3, 4)

print("a:\n", a)
print("b:\n", b)
print("c:", c)
print("d:\n", d)
print("d.shape:", d.shape)
```

```python
revenue = np.array([
    [100, 120,  90],
    [ 80, 110, 130],
    [150, 140, 160]
])  # shape = (3, 3)

day_adjust = np.array([10, -5, 0])   # shape = (3,)

adjusted_by_day = revenue + day_adjust

print("revenue:\n", revenue)
print("day_adjust:", day_adjust)
print("adjusted_by_day:\n", adjusted_by_day)
print("shapes:", revenue.shape, day_adjust.shape, adjusted_by_day.shape)


store_bonus = np.array([5, 0, 10]).reshape(3, 1)   # shape = (3,1)

adjusted_by_store = revenue + store_bonus

print("store_bonus:\n", store_bonus)
print("adjusted_by_store:\n", adjusted_by_store)
print("shapes:", revenue.shape, store_bonus.shape)
```

```python
revenue = np.array([
    [100, 120,  90],
    [ 80, 110, 130],
    [150, 140, 160]
])

col_sum = revenue.sum(axis=0)   # tổng từng cột
row_sum = revenue.sum(axis=1)   # tổng từng hàng
row_mean = revenue.mean(axis=1)

print("col_sum (axis=0):", col_sum)
print("row_sum (axis=1):", row_sum)
print("row_mean (axis=1):", row_mean)
```
