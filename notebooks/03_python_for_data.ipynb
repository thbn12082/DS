```python
cities = ["Ha Noi", "Da Nang", "Ha Noi", "HCM", "Da Nang"]
unique_cities = set(cities)

print(unique_cities)
print(len(unique_cities))
```

```python
def unique_keep_order(items):
    seen = set()
    result = []
    for x in items:
        if x not in seen:
            seen.add(x)
            result.append(x)
    return result

print(unique_keep_order(cities))
```

```python
sales = [
    {"product": "A", "revenue": 120, "region": "north"},
    {"product": "B", "revenue": 80,  "region": "south"},
    {"product": "C", "revenue": 200, "region": "north"},
    {"product": "D", "revenue": 50,  "region": "north"},
]

def top_k_products_by_region(data, region, k=2):
    # 1) filter
    filtered = []
    for row in data:
        if row["region"] == region:
            filtered.append(row)

    # 2) sort desc theo revenue
    ranked = sorted(filtered, key=lambda x: x["revenue"], reverse=True)

    # 3) top-k
    return ranked[:k]

print(top_k_products_by_region(sales, "north", 2))
```
