```python
scores = np.array([7, 8, 9, 10, 6])

mask = scores >= 8
print("mask:", mask)
print("filtered:", scores[mask])
```

```python
print(scores[scores >= 8])
```

```python
print(scores[scores % 2 == 0])
```

```python
scores = np.array([7, 8, 9, 10, 6])

# 1) cộng bonus 0.5 điểm
bonus_scores = scores + 0.5

# 2) lọc các bạn >= 8.5 sau bonus
passed = bonus_scores[bonus_scores >= 8.5]

# 3) thống kê
print("bonus_scores:", bonus_scores)
print("passed:", passed)
print("pass_count:", passed.size)
print("pass_mean:", passed.mean())
```
