```python
def get_even_numbers(nums):
    result = []
    for n in nums:
        if n % 2 == 0:
            result.append(n)
    return result

print(get_even_numbers([1, 2, 3, 4, 5, 6]))
print(get_even_numbers([]))          # []
print(get_even_numbers([1, 3, 5]))   # []
```

```python
print(get_even_numbers([]))          # []
print(get_even_numbers([1, 3, 5,2]))   # []
```

```python
def normalize_names(names):
    result = []
    for name in names:
        clean = name.strip().lower()
        result.append(clean)
    return result

print(normalize_names(["  An ", "BÌNH", "  cHuNg  "]))
```

```python
def safe_mean(nums):
    if len(nums) == 0:
        return None
    return sum(nums) / len(nums)

print(safe_mean([10, 20, 30]))
print(safe_mean([]))
```

```python
def word_freq(text):
    freq = {}
    words = text.lower().split()  # split() tự xử lý nhiều khoảng trắng
    for w in words:
        freq[w] = freq.get(w, 0) + 1
    return freq

print(word_freq("data science data  AI ai"))
```

```python
def split_pass_fail(scores, threshold=5):
    result = {"pass": [], "fail": []}
    for s in scores:
        if s >= threshold:
            result["pass"].append(s)
        else:
            result["fail"].append(s)
    return result

print(split_pass_fail([3, 5, 8, 4, 10]))
```
